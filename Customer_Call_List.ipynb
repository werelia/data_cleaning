```python
import pandas as pd
```

```python
df = pd.read_excel("Customer Call List.xlsx")
df
```

I clean data to get a list of clients.

I remove the missing data because they don't provide any value.

```python
df = df.dropna()
df
```

I standardize the column names to maintain consistency.

```python
df=df.rename(columns = {'CustomerID':'Customer_ID', 'Paying Customer':'Paying_Customer'})
df
```

I remove duplicates.

```python
df.drop_duplicates()
```

I remove unusful columns.

```python
df = df.drop(columns='Not_Useful_Column')
df
```

I standardize the notation of individual columns.

```python
df['Last_Name']=df['Last_Name'].str.strip('/._')
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')

df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))

df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
df[['Street_Adress', 'State', 'Zip_Code']]=df['Address'].str.split(',',2, expand=True)
df
```

```python
df['Paying_Customer']=df['Paying_Customer'].str.replace('Yes', 'Y')
df['Paying_Customer']=df['Paying_Customer'].str.replace('No', 'N')

df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No', 'N')

df
```

```python
df = df.replace('N/a','')
df
```

I remove clients from the database who do not wish to be contacted.

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

I reset the index.

```python
df = df.reset_index(drop=True)
df
```
